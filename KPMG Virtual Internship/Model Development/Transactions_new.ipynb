```python
table1 = pd.DataFrame(transactions_new.groupby('customer_id').recent.min())
```

```python
table2 = pd.DataFrame(transactions_new.groupby('customer_id').product_id.count())
```

```python
table2.head()
```

```python
table3 = pd.DataFrame(transactions_new.groupby('customer_id').profit.sum())
```

```python
table3.head()
```

```python
table4 = pd.merge(table1,table2, left_index = True, right_index =True)
```

```python
table4
```

```python
new_table = pd.merge(table4,table3,left_index = True, right_index = True)
```

```python
new_table.head()
```

```python
new_table.rename(columns = {'recent':'min_recent','product_id':'count_product_id','profit':'sum_profit'},inplace = True)
```

```python
new_table.head()
```

```python
new_table['min_recent'] = new_table.min_recent.dt.days
```

```python
new_table['min_recent'].describe()
```

```python
new_table['count_product_id'].describe()
```

```python
new_table['sum_profit'].describe()
```

```python
#define bins
bins1 = [0,18,45,88,353]
group_names1 = [4,3,2,1]
```

```python
new_table['recent_score'] = pd.cut(new_table['min_recent'],bins1,labels= group_names1)
```

```python
bins2 = [1,4,5,7,14]
group_names2 = [1,2,3,4]
```

```python
new_table['frequency_score'] = pd.cut(new_table['count_product_id'], bins2, labels= group_names2)
```

```python
bins3 = [15,1768,2786,4064,11669]
group_names3 = [1,2,3,4]
```

```python
new_table['profit_score'] = pd.cut(new_table['sum_profit'], bins3, labels= group_names3)
```

```python
new_table.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#to convert categorical values into numeric values
new_table['recent_score_new'] = LabelEncoder().fit_transform(new_table['recent_score'])
```

```python
new_table.head()
```

```python
new_table = new_table[new_table['recent_score_new']<=3]
```

```python
#in order to get the original values
new_table['recent_score_new'] = new_table['recent_score_new'] + 1
```

```python
new_table['frequency_score_new'] = LabelEncoder().fit_transform(new_table['frequency_score'])
```

```python
new_table = new_table[new_table['frequency_score_new']<=3]
```

```python
#in order to get the original values
new_table['frequency_score_new'] = new_table['frequency_score_new'] + 1
```

```python
new_table['profit_score_new'] = LabelEncoder().fit_transform(new_table['profit_score'])
```

```python
new_table['profit_score_new'].value_counts()
```

```python
#in order to get the original values
new_table['profit_score_new'] = new_table['profit_score_new'] + 1
```

```python
new_table.drop(columns = ['recent_score', 'frequency_score', 'profit_score'], inplace =True)
```

```python
#Giving more weightage to recency, then frequency and then to profit
new_table['total_score']= new_table['recent_score_new']*100+ new_table['frequency_score_new']*10+ new_table['profit_score_new']
```
